```python
# Criando uma função para o df acima
def categorizar_tomate_media(media): # recebe o valor da média
    if media >= 40 and media <= 70:
        return 'tomate medio'
    elif media < 40:
        return 'tomate pequeno'
    else:
        return 'tomatão'
```

```python
# Aplicando a função e criando a coluna 'Categoria_Tomate'
df_csv['Categoria_Tomate'] = df_csv['Average'].apply(categorizar_tomate_media)
```

### Alguns Conceitos - Preparar Base para ML

```python
# Primeiro conceito: é necessário separar a base / Preciso testar uma base que a Máquina nunca viu
# Preciso de uma base de treino e de uma base de teste
# Por este motivo separamos as bases

```

```python
# Selecionando uma parte da biblioteca para trabalhar
from sklearn.model_selection import train_test_split
```

```python
# Separando os dados em variáveis x e y
# No x vou colocar toda a minha base, execeto a última coluna (tb poderia colocar só algumas colunas se eu quisesse)

x = df_diabetes.drop(['Class variable'], axis=1)
```

```python
x.head()
```

```python
# A coluna 'Class variable' será a variável y
y = df_diabetes['Class variable']
```

```python
y
```

```python
# Separando a base em treino e teste
# em train_test_split temos que definir quantos % será de teste e treino. No caso, teste será 30% então 70% é para treino / a maior % deve ser para treino.
# importante comentar que a divisão sõ realizadas aleatoriamente, ou seja, pega 70% de linhas aleatórias e não sequenciais
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# len: conta quantas linhas tenho no df / / foram selecionadas linhas aleatórias
len(x_train)
```

```python
# de 768 linhas, 537 foram separadas para a base de treino
df_diabetes.shape
```

```python
# 231 linhas para a base de teste / foram selecionadas linhas aleatórias
len(x_test)
```

### Montando o primeiro Algorítimo

```python
# Com o from, pegamos parte da Biblioteca / da biblioteca sklearn quero somente o KneighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Qual a quantidade de vizinhos que devem ser reconhecidos para classificar os dados
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Treinando o algoratímo
knn.fit(x_train, y_train)
```

```python
# Depois da criação do algorítimo de treino, é necessário testá-lo: Acurácia do Modelo
accuracy = knn.score(x_test, y_test)
```

```python
accuracy
```

### Entendendo a Base de Dados

As colunas neste conjunto de dados são:

nome: o nome da música 
 
álbum: o nome do albúm  

release_date: o dia, mês e ano em que um albúm foi lançado  

número de faixa: a ordem em que a música aparece no albúm  

id: o id do Spotify para a música  

uri: o uri do Spotify para a música (url)  

acústica: Uma medida de confiança de 0,0 a 1,0 se a faixa é acústica. 1,0 representa alta confiança de que a faixa é acústica.  

danceability: Danceability descreve o quanto uma faixa é adequada para dançar com base em uma combinação de elementos musicais, incluindo andamento, estabilidade do rítimo, força da batida e regularidade geral. Um valor de 0,0 é menos dançável e 1,0 é o mais dançável.  

energia: A energia é uma medida de 0,0 a 1,0 e representa uma medida perceptiva de intensidade e atividade. Normalmente, as faixas energéticas parecem rápidas, altas e barulhentas. Por exemplo, death metal tem alta energia, enquanto prelúdio de Batch pontua baixo na escala. Os recursos perceptíveis que contribuem para esse atributo incluem faixa dinâmica, sonoridade percebida, timbre, taxa de início e entropia geral.
