```python
# Criando uma função para o df acima
def categorizar_tomate_media(media): # recebe o valor da média
    if media >= 40 and media <= 70:
        return 'tomate medio'
    elif media < 40:
        return 'tomate pequeno'
    else:
        return 'tomatão'
```

```python
# Aplicando a função e criando a coluna 'Categoria_Tomate'
df_csv['Categoria_Tomate'] = df_csv['Average'].apply(categorizar_tomate_media)
```

```python
# Visualizando os dados com a aplicação da função
df_csv.head()
```

```python
# Agrupando por categoria 'tomate'
df_csv.groupby(['Categoria_Tomate']).describe()
```

```python
# Filtrando Dados / Localizando Dados
filtro = df_csv['Average'] < 40
df_csv.loc[filtro]
```
